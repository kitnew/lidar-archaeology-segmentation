```python
import numpy as np

file_structure = {
    "data": np.float32,   # [C, H, W]
    "mask": np.uint8,     # [H, W]
    "soft_mask": np.float32, # [H, W]
    "valid": np.uint8,    # [H, W]
    "meta": {
        "height": int,
        "width": int,
        "nodata_fill_value": float,
        "mask_line_width_m": 1.0,
        "mask_buffer_distance_m": 0.5,
        "pixel_size_m": 0.25,
        "created_from": {
            "data_path": str,
            "mask_path": str
        }
    }
}

no_data_value = 0.0
```

```python
def zscore(x, mask):
    # x: [C, H, W], mask: [H, W]
    for c in range(x.shape[0]):
        valid_pixels = x[c][mask]
        if valid_pixels.size > 0:
            m = valid_pixels.mean()
            s = valid_pixels.std()
            x[c] = (x[c] - m) / (s + 1e-8)
    x[:, ~mask] = no_data_value
    return x

def minmax(x, mask):
    for c in range(x.shape[0]):
        valid_pixels = x[c][mask]
        if valid_pixels.size > 0:
            mi = valid_pixels.min()
            ma = valid_pixels.max()
            if ma > mi:
                x[c] = (x[c] - mi) / (ma - mi)
    x[:, ~mask] = no_data_value
    return x
```

```python
import os
import numpy as np
import rasterio

# -----------------------------
# OUTPUT DIR
# -----------------------------
out_dir = "/home/nikitachernysh/Storage/Projects/lidar-archaeology-segmentation/thesis/data/processed/terrases"
os.makedirs(out_dir, exist_ok=True)

# -----------------------------
# CONSTANTS
# -----------------------------
no_data_value = 0.0

# -----------------------------
# MAIN LOOP
# -----------------------------
for file_path, mask_path, out_name in zip(files, masks, outfile_names):
    print(f"\nProcessing: {out_name}")

    # --- verify file doesnt exist ---

    out_path = os.path.join(out_dir, f"{out_name}.npz")

    if os.path.isfile(out_path):
        print("File already exists ... Skip")
        continue

    # --- read input raster ---
    with rasterio.open(file_path) as src:
        data = src.read().astype(np.float32)   # [C, H, W]
        height = src.height
        width = src.width
        raster_nodata = src.nodata

    # --- build valid mask ---
    # Если nodata в raster задан, используем его.
    # Если нет, считаем всё valid.
    if raster_nodata is not None:
        if data.shape[0] == 1:
            valid = (data[0] != raster_nodata).astype(np.uint8)
        else:
            valid = np.all(data != raster_nodata, axis=0).astype(np.uint8)
    else:
        valid = np.ones((height, width), dtype=np.uint8)

    # --- replace invalid pixels with no_data_value ---
    if data.shape[0] == 1:
        data[0][valid == 0] = no_data_value
    else:
        for c in range(data.shape[0]):
            data[c][valid == 0] = no_data_value

    # --- normalize DEM --- 
    method = out_name.split('_')[-2] if out_name.find("_") != -1 else ""
    if method == "znorm":
        data = zscore(data, valid.astype(bool))
    elif method == "minmaxnorm":
        data = minmax(data, valid.astype(bool))

    # --- remove or stack channels

    if data.shape[0] > 3:
        data = data[0:3]
    if data.shape[0] == 1:
        data = np.stack([data[0], data[0], data[0]])

    # --- read mask or create empty mask ---
    if mask_path is not None:
        with rasterio.open(mask_path) as msrc:
            mask = msrc.read(1)

            if msrc.height != height or msrc.width != width:
                raise ValueError(
                    f"Mask shape mismatch for {out_name}: "
                    f"mask=({msrc.height}, {msrc.width}) vs data=({height}, {width})"
                )

        mask = (mask > 0).astype(np.uint8)
    else:
        mask = np.zeros((height, width), dtype=np.uint8)

    # --- optional: zero mask outside valid ---
    mask[valid == 0] = 0

    # --- soft mask ---
    #soft_mask = np.zeros((height, width), dtype=np.float32)
    import numpy as np
    from scipy.ndimage import distance_transform_edt

    # input: mask -> 2D numpy array, binary labeling mask

    radius = 2
    resolution = 0.25  # meters per pixel

    # paper values in meters
    shadow_len_m = radius

    shadow_len_px = shadow_len_m / resolution

    mask = (mask > 0).astype(np.uint8)

    dist_outside = distance_transform_edt(mask == 0)

    soft_mask = np.where(
        mask == 1,
        1.0,
        np.clip(1.0 - dist_outside / shadow_len_px, 0.0, 1.0)  # pyright: ignore[reportOperatorIssue]
    ).astype(np.float32)

    # --- metadata ---
    meta = {
        "height": int(height),
        "width": int(width),
        "nodata_fill_value": float(no_data_value),
        "mask_line_width_m": 1.0,
        "mask_buffer_distance_m": 0.5,
        "pixel_size_m": 0.25,
        "created_from": {
            "data_path": file_path,
            "mask_path": mask_path if mask_path is not None else "None"
        }
    }

    file_structure["data"] = data
    file_structure["mask"] = mask
    file_structure["soft_mask"] = soft_mask
    file_structure["valid"] = valid

    file_structure["meta"]["height"] = data[0].shape[0]
    file_structure["meta"]["width"] = data[0].shape[1]
    file_structure["meta"]["nodata_fill_value"] = no_data_value

    file_structure["meta"] = meta

    # --- save ---

    np.savez_compressed(out_path, **file_structure)

    print(f"Saved: {out_path}")
    print(f"data shape: {data.shape}")
    print(f"mask shape: {mask.shape}")
    print(f"valid shape: {valid.shape}")
    print(f"positive pixels: {int(mask.sum())}")
    print(f"valid pixels: {int(valid.sum())}")
```
